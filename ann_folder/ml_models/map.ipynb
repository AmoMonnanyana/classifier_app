```python


def generate_map(shape, set):
    dataset =pd.read_excel(set)
    geometry = [Point(xy) for xy in zip(dataset['Longitude'], dataset['Latitude'])]
    gdf = gpd.GeoDataFrame(dataset, geometry=geometry)
    gdf.crs = "EPSG:4326"
    boundaries = gpd.read_file(shape)
    base = boundaries.plot(figsize =(20, 20), color = 'white', aspect=1, edgecolor='black')
    gdf.plot(ax=base,markersize=gdf['Cr (mg/kg)'], column='Cr (mg/kg)', legend=True)
    return "successful"
```

```python
set = "C:/Users/amoge/Downloads/Soil_dataset.xlsx"
shapefile = "C:/Users/amoge/Downloads/gadm41_ZAF_shp/gadm41_ZAF_1.shp"

generate_map(shapefile, set)
```
